```python
#  Import Libraries import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load Dataset (assuming CSV in same directory)
df = pd.read_csv('titanic.csv')
# First few records
df.head()

```

```python
#1. Survival Count
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
# 2. Survival by Gender
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Gender')
plt.show()
```

```python
#3. Passenger Class Distribution (Pie Chart)
df['Pclass'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Passenger Class Distribution')
plt.ylabel('')
plt.show()
```

```python
#4. Age Distribution (Histogram)
plt.hist(df['Age'].dropna(), bins=30, color='skyblue')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
#5. Age vs Passenger Class (Boxplot)
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title('Age Distribution Across Passenger Classes')
plt.show()
```

```python
#6. Heatmap for Correlation Matrix
# Select numeric columns only
numeric_df = df.select_dtypes(include=['number'])

# Heatmap
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()

```

```python
#7. Pairplot for key variables
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass']])
plt.show()
```

```python
#8. Fare Distribution by Passenger Class
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title('Fare Distribution Across Passenger Classes')
plt.show()
```

```python
#9. Survival by Embarked Location
sns.countplot(x='Embarked', hue='Survived', data=df)
plt.title('Survival by Embarked Port')
plt.show()
```

```python
#10. Family Size vs Survival
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
sns.barplot(x='FamilySize', y='Survived', data=df)
plt.title('Survival Rate by Family Size')
plt.show()
```
